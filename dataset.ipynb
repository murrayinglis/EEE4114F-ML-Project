```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python

# Paths to your images
image_paths = ["Data/circle1.jpg", "Data/square1.jpg", "Data/triangle6.jpg", "Data/pentagon1.jpg", "Data/hexagon1.jpg"]

# Create a subplot grid
fig, axes = plt.subplots(1, 5, figsize=(15, 3))

# Iterate over images and place them on the plot
for i, path in enumerate(image_paths):
    img = mpimg.imread(path)
    axes[i].imshow(img)
    axes[i].axis('off')  # Turn off axis

axes[0].set_title("Circle")
axes[1].set_title("Square")
axes[2].set_title("Triangle")
axes[3].set_title("Pentagon")
axes[4].set_title("Hexagon")

# Show the plot
plt.tight_layout()
plt.show()
```
