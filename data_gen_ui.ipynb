```python
import tkinter as tk
from tkinter import Canvas, Button
from PIL import Image, ImageDraw, ImageOps
import os
import random
import cv2
```

# Drawing data

Circles

```python
class DrawingApp:
    def __init__(self, master, width=240, height=240):
        self.master = master
        self.width = width
        self.height = height
        
        self.find_highest_image_count()  # Find the highest numbered image file
        self.canvas = tk.Canvas(master, width=self.width, height=self.height, bg="white")
        self.canvas.pack()
        
        self.canvas.bind("<B1-Motion>", self.draw)
        
        self.button_save = tk.Button(master, text="Save", command=self.save_image)
        self.button_save.pack()
        
        self.new_canvas()  # Create a new drawing canvas
        
    def draw(self, event):
        x, y = event.x, event.y
        r = 3
        self.canvas.create_oval(x-r, y-r, x+r, y+r, fill="black")
        self.draw.ellipse([x-r, y-r, x+r, y+r], fill="black")
        
    def save_image(self):
        directory = "Shapes"
        if not os.path.exists(directory):
            os.makedirs(directory)
        filename = f"{directory}/circle{self.image_count}.png"
        self.image.save(filename)
        print(f"Image saved as {filename}")
        self.image_count += 1
        self.new_canvas()  # Create a new drawing canvas
        
    def new_canvas(self):
        self.canvas.delete("all")  # Clear the canvas
        self.image = Image.new("RGB", (self.width, self.height), "white")
        self.draw = ImageDraw.Draw(self.image)
        
    def find_highest_image_count(self):
        image_files = [f for f in os.listdir("Shapes") if f.startswith("circle") and f.endswith(".png")]
        if image_files:
            self.image_count = max(int(f.split("circle")[1].split(".png")[0]) for f in image_files) + 1
        else:
            self.image_count = 1

def main():
    root = tk.Tk()
    root.title("CIRCLES")
    app = DrawingApp(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```

Rotations

```python
def apply_random_transformation(input_file, output_file):
    # Open the input image
    with Image.open(input_file) as img:
        # Apply random rotation (-10 to 10 degrees)
        angle = random.uniform(-20, 20)
        rotated_img = img.rotate(angle, resample=Image.BICUBIC, expand=True)
        
        # Get the bounding box after rotation
        bbox = rotated_img.getbbox()
        
        # Crop the rotated image to fit within the bounding box
        cropped_img = rotated_img.crop(bbox)
        
        # Create a new white background image
        background = Image.new('RGB', img.size, 'white')
        
        # Paste the cropped image onto the white background
        background.paste(cropped_img, (0,0))
        
        # Resize the image to 240x240
        background = background.resize((240, 240), resample=Image.BICUBIC)
        
        # Save the transformed image
        background.save(output_file)
        print(f"Image saved as {output_file}")


# Example usage:
input_file = "Shapes/circle1.png"
output_file = "Data/circle1.png"
apply_random_transformation(input_file, output_file)
```
